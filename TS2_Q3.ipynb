```python
Sut = 210
Suc = 630
```

```python
from math import sqrt
```

## Part (a)

There are no shear stresses. So the given normal stresses are themselves the principal stresses. 

$\sigma_A = 175$ and $\sigma_B = 105$

Here, $\sigma_A > \sigma_B > 0$. 

So, both the Brittle Coulomb-Mohr and the modified Mohr failure criteria to be used are:
$\displaystyle \frac{\sigma_A}{S_{ut}} - \frac{0}{S_{uc}} = \frac{1}{n}$

```python
sigma_A = 175
n_BCM = Sut/sigma_A

n_MM = Sut/sigma_A

display(n_BCM, n_MM)
```

## Part (b)

Again, there are no shear stresses. So the given normal stresses are themselves the principal stresses. 

$\sigma_A = 105$ and $\sigma_B = -105$. 

Here, $\sigma_A > 0 > \sigma_B$. 

Brittle Coulomb-Mohr: $\displaystyle \frac{\sigma_A}{S_{ut}} - \frac{\sigma_B}{S_{uc}} = \frac{1}{n}$

Modified Coulomb-Mohr: Since $\left|\dfrac{\sigma_B}{\sigma_A}\right| = 1$, Case 2 works: $\displaystyle \frac{\sigma_A}{S_{ut}} = \frac{1}{n}$

```python
sigma_A = 105
sigma_B = -105

n_BCM = (sigma_A/Sut - sigma_B/Suc)**(-1)
n_MM = Sut/sigma_A

display(n_BCM, n_MM)
```

## Part (c)

This time there is shear stress. So, first the principal stresses have to be found. 

$\displaystyle \sigma_{A,B} = \frac{\sigma_x + \sigma_y}{2} \pm \sqrt{\left( \frac{\sigma_x - \sigma_y}{2} \right)^2 + \tau_{xy}^2}$

```python
sigma_x = 140
sigma_y = 0
tau_xy = -70

sigma_A = (sigma_x + sigma_y)/2 + sqrt( ((sigma_x - sigma_y)/2)**2 + tau_xy**2 )
sigma_B = (sigma_x + sigma_y)/2 - sqrt( ((sigma_x - sigma_y)/2)**2 + tau_xy**2 )
display(sigma_A, sigma_B)
```

Here, $\sigma_A > 0 > \sigma_B$. 

Brittle Coulomb-Mohr: $\displaystyle \frac{\sigma_A}{S_{ut}} - \frac{\sigma_B}{S_{uc}} = \frac{1}{n}$

Modified Coulomb-Mohr: Since $\left|\dfrac{\sigma_B}{\sigma_A}\right| < 1$, Case 2 works: $\displaystyle \frac{\sigma_A}{S_{ut}} = \frac{1}{n}$

```python
n_BCM = (sigma_A/Sut - sigma_B/Suc)**(-1)
n_MM = Sut/sigma_A

display(n_BCM, n_MM)
```

## Part (d)

We first find the principal stresses:

```python
sigma_x = -105
sigma_y = -70
tau_xy = -105

sigma_A = (sigma_x + sigma_y)/2 + sqrt( ((sigma_x - sigma_y)/2)**2 + tau_xy**2 )
sigma_B = (sigma_x + sigma_y)/2 - sqrt( ((sigma_x - sigma_y)/2)**2 + tau_xy**2 )
display(sigma_A, sigma_B)
```

Here, $\sigma_A > 0 > \sigma_B$. 

Brittle Coulomb-Mohr: $\displaystyle \frac{\sigma_A}{S_{ut}} - \frac{\sigma_B}{S_{uc}} = \frac{1}{n}$

Modified Coulomb-Mohr: Since $\left|\dfrac{\sigma_B}{\sigma_A}\right| > 1$, Case 3 works: 
$\displaystyle \frac{(S_{uc}-S_{ut})\sigma_A}{S_{uc} S_{ut}} - \frac{\sigma_B}{S_{uc}} = \frac{1}{n}$

```python
n_BCM = (sigma_A/Sut - sigma_B/Suc)**(-1)
n_MM = ( (Suc-Sut)*sigma_A/(Suc*Sut) - sigma_B/Suc )**(-1)

display(n_BCM, n_MM)
```

## Part (e)

Again, there are no shear stresses. So, the given normal stresses are themselves the principal stresses. 

```python
sigma_A = -84
sigma_B = 105
```

Here, $\sigma_B > 0 > \sigma_A$ 

Since $\sigma_B > \sigma_A$, we have to use the counterpart to the cases discussed in class where $\sigma_A > \sigma_B$. 

Brittle Coulomb-Mohr: $\displaystyle \frac{\sigma_B}{S_{ut}} - \frac{\sigma_A}{S_{uc}} = \frac{1}{n}$ 

Modified Coulomb-Mohr: Since $\displaystyle \left| \frac{\sigma_A}{\sigma_B}\right|<1$, the counterpart to Case 2 works:
$\displaystyle \frac{\sigma_B}{S_{ut}} = \frac{1}{n}$

```python
n_BCM = (sigma_B/Sut - sigma_A/Suc)**(-1)

n_MM = Sut/sigma_B

display(n_BCM, n_MM)
```
