```python
from math import pi, sqrt
```

## In Imperial units

```python
Sut = 55 #380 MPa
Sy = 30 #210 MPa

Kts = 1.6
L = 2

Fmin = 150
Fmax = 500
qs = 0.8

Kfs = 1 + qs*(Kts - 1)

Tmax = 500*2
Tmin = 150*2

d = 7/8

tau_max = Kfs*16*Tmax/(pi*d**3)/1e3
tau_min = Kfs*16*Tmin/(pi*d**3)/1e3

tau_m = (tau_max + tau_min)/2
tau_a = (tau_max - tau_min)/2

Se_dash = 0.5*Sut
ka = 11.0*(Sut)**(-0.650)
de = 0.370*d
kb = 0.879*(de)**(-0.107)
kc = 0.59

Se = ka*kb*kc*Se_dash

Ssu = 0.67*Sut

#Modified Goodman
nf_MG = 1/(tau_a/Se + tau_m/Ssu)
display(nf_MG)

#Gerber
nf_Gerber = 1/2*(Ssu/tau_m)**2*(tau_a/Se)*(-1+sqrt(1+(2*tau_m*Se/(Ssu*tau_a))**2))
display(nf_Gerber)
```

## In SI units

```python
Sut = 380
Sy = 210 

Kts = 1.6
L = 1

Fmin = 0.5e3
Fmax = 2e3
qs = 0.8

Kfs = 1 + qs*(Kts - 1)

Tmax = Fmax*2*25
Tmin = Fmin*2*25

d = 20

tau_max = Kfs*16*Tmax/(pi*d**3)
tau_min = Kfs*16*Tmin/(pi*d**3)

tau_m = (tau_max + tau_min)/2
tau_a = (tau_max - tau_min)/2

Se_dash = 0.5*Sut
ka = 38.6*(Sut)**(-0.650)
de = 0.370*d  #Eq. 6-24
kb = 1.24*(de)**(-0.107)  #Eq. 6-20, third line
kc = 0.59  #Eq. 6-26

Se = ka*kb*kc*Se_dash

Ssu = 0.67*Sut  #Eq. 6-54

#Modified Goodman
nf_MG = 1/(tau_a/Se + tau_m/Ssu)
display(nf_MG)

#Gerber
nf_Gerber = 1/2*(Ssu/tau_m)**2*(tau_a/Se)*(-1+sqrt(1+(2*tau_m*Se/(Ssu*tau_a))**2))
display(nf_Gerber)
```
