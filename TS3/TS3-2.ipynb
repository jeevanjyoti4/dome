Note that for low-cycle fatigue, the stress-life method is not recommended. We will check if the stress corresponding to the 500 cycles is at least low enough to not cause yielding. 

We will be in the first segment of the S-N diagram. The operating point will lie on this line segment joining the two points: ($10^0, S_{ut}$) and ($10^3, fS_{ut}$). 

The equation of this line in the log-log S-N diagram is: $S_f = a N^b$, where $a = S_{ut}$ and $b = \dfrac{1}{3} \log f$. 

So, first of all we need to determine $f$ from Fig. 6-23 (Shigley, 10th edition). 

```python
from math import log10
```

```python
f = 0.80 #eye-estimate from the plot of Fig. 6-23

a = Sut
b = 1/3*log10(f)

display(a, b)
```

```python
N = 500
Sf = a*N**b
display(Sf)
```

Since $S_f < S_y$, it is possible to do the stress-life test at the low cycle life of $N = 500$.
