```python
Sut = 590 # From Table A-20
Sy = 490 # From Table A-20

Fmin = 0e3
Fmax = 28e3

D = 6
w = 25
t = 10

A = (w-D)*t

sigma_max = Fmax/A

ny = Sy/sigma_max
display(ny)
```

We will now determine the fatigue factor of safety based on infinite life. 

```python
Se_dash = 0.5*Sut  # Eq. 6-8
ka = 3.04*Sut**(-0.217)   # Eq. 6-19 and Table 6-2, corresponding to cold-drawn
kb = 1 # Eq. 6-21, i.e. for axial load
kc = 0.85 # Eq. 6-26, corresponding to axial load

Se = ka*kb*kc*Se_dash
display(Se)
```

```python
q = 0.83 # Fig. 6-20

Kt = 2.44 # From Fig. A-15-1

Kf = 1 + q*(Kt - 1)
display(Kf)

sigma_a = Kf*abs((Fmax - Fmin)/2)/A
sigma_m = Kf*((Fmax+Fmin)/2)/A
display(sigma_a, sigma_m)

nf = (sigma_a/Se + sigma_m/Sut)**(-1)
display(nf)
```

Since nf is less than 1, it is not a case of infinite life. So, we have to estimate the life from the S-N diagram. 

However, we have to first determine an equivalent completely reversed stress. This equivalent completely reversed stress is basically the intercept on the ordinate of the line that passes through (sigma_a, sigma_m) and (0, Sut). 

```python
sigma_eq_rev = sigma_a/(1 - sigma_m/Sut)
display(sigma_eq_rev)
```

```python
f = 0.87  # From Fig. 6-18

a = (f*Sut)**2/Se # Eq. 6-14

b = -1/3*log10(f*Sut/Se)

N = (sigma_a/a)**(1/b)

display(N)
```
