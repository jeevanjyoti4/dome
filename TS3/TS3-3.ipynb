```python
Se = 276
Sy = 413
Sut = 551
```

```python
sigma_a = 172
sigma_m = 0
tau_a = 0
tau_m = 103
```

## Factor of safety against static failure:

We have to first determine the maximum stresses by adding the amplitude of the stresses to the mean (or, mid-range) stresses:

```python
sigma_max = sigma_m + sigma_a
tau_max = tau_m + tau_a
display(sigma_max, tau_max)
```

Next, we determine the von Mises stress corresponding to these maximum stresses:

```python
sigma_vM_max = (sigma_max**2 + 3*tau_max**2)**(1/2)
```

Finally, we can obtain the factor of safety against static failure:

```python
n_s = Sy/sigma_vM_max
display(n_s)
```

## For fatigue failure:

We have to first determine the von Mises stress corresponding to each of the alternating and mean stresses:

```python
sigma_vM_a = (sigma_a**2 + 3*tau_a**2)**(1/2)

sigma_vM_m = (sigma_m**2 + 3*tau_m**2)**(1/2)
display(sigma_vM_a, sigma_vM_m)
```

We next check for the factor of safety for each of the (a) Goodman, (b) Gerber, and (c) ASME-elliptic criteria. 

### Goodman
$n_f = \left( \dfrac{\sigma_a}{S_e} + \dfrac{\sigma_m}{S_{ut}} \right)^{-1}$

```python
nf_Goodman = (sigma_vM_a/Se + sigma_vM_m/Sut)**(-1)
display(nf_Goodman)
```

### Gerber
$n_f = \dfrac{1}{2} \left( \dfrac{S_{ut}}{\sigma_m}\right)^2 \left(  \dfrac{\sigma_a}{S_e} \right) \left[ -1 + \sqrt{ 1 + \left( \dfrac{2 \sigma_m S_e}{\sigma_a S_{ut}} \right)^2 } \right]$

```python
nf_Gerber = 1/2*(Sut/sigma_vM_m)**2*(sigma_vM_a/Se)*(-1 + (1 + (2*sigma_vM_m*Se/(sigma_vM_a*Sut))**2)**(1/2))
display(nf_Gerber)
```

### ASME-elliptic
$n_f = \left[ \left( \dfrac{\sigma_a}{S_e} \right)^2 + \left( \dfrac{\sigma_m}{S_y} \right)^2 \right]^{-1/2}$

```python
nf_ASMEelliptic = ( (sigma_vM_a/Se)**2 + (sigma_vM_m/Sy)**2 )**(-1/2)
display(nf_ASMEelliptic)
```

**VERY IMPORTANT**: Note that the factor of safety in each of the fatigue failure criteria is greater than 1. So the operating point is safely within the infinite life zone. Had the factor of safety been less than 1, then the operating point would have been in the finite life zone, and then it would be necessary to determine the expected life as mentioned in the question.
