From Table 11-2, we observe that for an 02-30 mm angular contact ball bearing, $C_{10} = 20.3$ kN. 



We can then determine the lives $L_1$ and $L_2$ corresponding to a radial load of $F_1 = 18$ kN and $F_2 = 30$ kN, respectively. 

```python
a = 3 #ball bearing

C10 = 20.3
L10 = 1e6


F1 = 18
F2 = 30

L1 = (C10/F1)**a*L10

L2 = (C10/F2)**a*L10
```

Now, the actual number of cycles at load $F_1$ is $l_1 = 200000$. 

We now use the following relation to obtain $l_2$:

\begin{gather*}
\frac{l_1}{L_1} + \frac{l_2}{L_2} = 1 \\
\implies l_2 = L_2 \left(1 - \frac{l_1}{L_1}\right)
\end{gather*}

```python
l1 = 200000

l2 = L2*(1-l1/L1)
display(l2)
```
