We have the relation:
    
\begin{gather*}
    C_{10} = F_D \left[ \frac{x_D}{x_0 + (\theta - x_0)[\ln(1/R_D) ]^{1/b}}  \right]^{1/a}
\end{gather*}

We have two different reliabilities: $R_{D1} = 0.90$ and $R_{D2} = 0.20$. 

We also have, respectively, two different values of life: $x_{D1}$ and $x_{D2}$. 

Also $x_0 = 0$. 

Therefore, we must have:
\begin{gather*}
\left[ \frac{x_{D1}}{\theta[\ln(1/R_{D1}) ]^{1/b}}  \right]^{1/a} = \left[ \frac{x_{D2}}{\theta[\ln(1/R_{D2}) ]^{1/b}}  \right]^{1/a} \\
\implies \frac{ [\ln(1/R_{D1}) ]^{1/b} }{ [\ln(1/R_{D2}) ]^{1/b} } = \frac{x_{D1}}{x_{D2}} \\
\implies \frac{1}{b} \log \left( \frac{\log R_{D1}}{\log R_{D2}} \right) = \frac{x_{D1}}{x_{D2}} \\
\implies b = \frac{\log \left( \frac{\log R_{D1}}{\log R_{D2}} \right)}{ \frac{x_{D1}}{x_{D2}} }
\end{gather*}


```python
RD1 = 0.90
RD2 = 0.20

xD1 = 115*2000*60/1e6  # rated at 1e6 revolutions
xD2 = 600*2000*60/1e6  # rated at 1e6 revolutions
```

```python
from math import log
```

```python
b = log(log(RD1)/log(RD2)) / log(xD1/xD2)
display(b)
```

For the calculation of $\theta$, we note that:
\begin{gather*}
\frac{10^6}{10^6} = x_0 + (\theta - x_0)(\log(1/0.90))^{1/b}
\end{gather*}

```python
theta = 1/(log(1/RD1))**(1/b)
display(theta)
```
