The important thing to note here is that the ball bearing is operating at 96 percent reliability. We have to "move" from the 96 percent reliability line to the 90 percent reliability line (at which the catalog rating is given). Refer to the figure below:




So, we need to use the following relation:
\begin{gather*}
    C_{10} = F_B = a_f F_D \left( \frac{x_D}{x_B} \right)^{1/a} = a_f F_D \left[ \frac{x_D}{x_0 + (\theta - x_0)[\ln(1/R_D) ]^{1/b}}  \right]^{1/a}
\end{gather*}
In the above expression, $x_D = L_D/L_{10}$, where $L_D$ is the life at point $D$, and $L_{10}$ is the life at point $A$.  

```python
from math import log
```

```python
x0 = 0
theta = 4.48
b = 1.5
RD = 0.96
FD = 2.75
af = 1.2  # application factor (acts like a factor of safety)
a = 3  # for ball bearings 

xB = x0 + (theta - x0)*(log(1/RD))**(1/b)

LD = 10000*1800*60
L10 = 3000*500*60
xD = LD/L10

C10 = af*FD*(xD/xB)**(1/a)
display(C10)
```

Note that the above answer does not match the answer given in the tutorial sheet. However, if we use the approximation, $\log(1/R_D) \approx 1 - R_D$, then:

```python
xB = x0 + (theta - x0)*(1-RD)**(1/b)
C10 = af*FD*(xD/xB)**(1/a)
display(C10)
```
