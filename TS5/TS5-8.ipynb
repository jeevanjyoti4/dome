We start by considering the following expression for equivalent force:

\begin{gather*}
    F_{\rm eq} = \left[ \frac{F_{e1}^a l_1 + F_{e2}^a l_2 + F_{e3}^a l_3}{l_1 + l_2 + l_3} \right]^{1/a} = \left[\sum f_i F_{ei}^a \right]^{1/a},
\end{gather*}

We know that:
\begin{gather*}
F_{\rm eq}^a L_{\rm eq} = F_{e1}^a L_1 = F_{e2}^a L_2 = F_{e3}^a L_3
\end{gather*}

Then the equivalent life will be: $L_{\rm eq} = \dfrac{K}{F_{\rm eq}} = \dfrac{K}{\left[\sum f_i F_{ei}^a \right]^{1/a}}$.

From Table 11-3 below, we start our trial with bore size of 25 mm. 



For the 25-mm bearing, $C_{10} = 28.6$ kN. 

We first determine the value of $K$ as $K = C_{10}^a L_{10}$.  

```python
C10 = 28.6
L10 = 1e6
a = 10/3 # roller bearing

K = C10**a*L10
```

```python
f1 = 0.30
f2 = 0.50
f3 = 0.20

Fe1 = 20
Fe2 = 25
Fe3 = 30

Leq = K/(f1*Fe1**a + f2*Fe2**a + f3*Fe3**a)
display(Leq)
```

The above $L_{\rm eq}$ is less than the desired $5\times10^6$ revolutions. So, we try a higher bore size. 

For the 30-mm bore size bearing, $C_{10} = 36.9$ kN.

We again calculate $K$ and then $L_{\rm eq}$. 

```python
C10 = 36.9

K = C10**a*L10

Leq = K/(f1*Fe1**a + f2*Fe2**a + f3*Fe3**a)
display(Leq)
```

The above $L_{\rm eq}$ is again less than the desired $5\times 10^6$ revolutions. So, we again try a higher bore size. 

For the 35-mm bore size bearing, $C_{10} = 44.6$ kN. 

As before, we calculate $K$ and then $L_{\rm eq}$. 

```python
C10 = 44.6

K = C10**a*L10

Leq = K/(f1*Fe1**a + f2*Fe2**a + f3*Fe3**a)
display(Leq)
```

This value of $L_{\rm eq}$ exceeds the desired $5 \times 10^6$ revolutions. And, so the minimum bore size that is to be selected is 35 mm.
