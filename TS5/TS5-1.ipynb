We use the load-life relation:
\begin{gather*}
F L^{1/a} = \text{constant (at a certain reliability)}
\end{gather*}

From the above relation, we have: $F_1 L_1^{1/a} = F_2 L_2^{1/a}$. 

```python
L1 = 3000*500*60 #converting to number of revolutions

L2 = 10000*1800*60 #converting to number of revolutions
F2 = 2.75

a = 3
```

```python
F1 = F2*(L2/L1)**(1/a)
display(F1)
```
