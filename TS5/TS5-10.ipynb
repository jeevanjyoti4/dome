From the solution of Q9, we note that:

```python
n = 2000 #rpm

a = 3 #ball bearing

C10 = 20.3
L10 = 1e6

F1 = 18
F2 = 30

L1 = (C10/F1)**a*L10

L2 = (C10/F2)**a*L10
```

Let the total number of cycles be $c$. Then we must have:
\begin{gather*}
\left( \frac{c n t_1}{L_1} + \frac{c n t_2}{L_2} \right) = 1
\end{gather*}
where $t_1 = 4$ min and $t_2 = 6$ min are the times spent in each 10 min cycle at the two loads of $F_1$ and $F_2$, respectively. 

```python
n = 2000 #rpm
t1 = 4
t2 = 6

c = 1/(n*t1/L1 + n*t2/L2)
display(c)
```

The total number of revolutions consisting of these 10 min cycles at 2000 rpm will be:

```python
L = c*10*n
display(L)
```

And, the total number of hours spent in the above number of revolutions will be:

```python
L/(60*2000)
```
