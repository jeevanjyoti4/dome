A comparison of load should be made corresponding to the same value of life. 

Here, for bearing $B$: $L_B = 10^6$ revolutions. 

Let's check the number of revolutions for $A$:

```python
L_A = 3000*500*60 
display(L_A)
```

That's $90\times10^6$ revolutions. If you look back at Problems 1 and 2 of this Tutorial sheet, you will note that bearing manufacturer must be Timken. 

Anyway, so we have to find the load that this Timken bearing can take at $10^6$ revolutions. Only then we will be able to compare with bearing $B$. 

```python
a = 3 # for ball bearings

C10 = 2 # the catalog rating for bearing A

F_A = C10*(L_A/1e6)**(1/a)  

display(F_A)
```

So, the load that bearing $A$ can take for a life of $10^6$ revolutions is 8.96 kN which is larger than the corresponding load of 7 kN for bearing $B$. 

Answer: Bearing $A$ can carry the larger load.
