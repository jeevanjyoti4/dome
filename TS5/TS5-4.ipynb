Let the angular contact bearing be designated by A and the cylindrical roller bearing by B. 

The overall reliability desired is $R = 0.90$.

Let us start out with equal reliability values for both the bearings. Thus, the desired reliability is: $R_D = \sqrt{R}$

```python
from math import sqrt
```

```python
R_D = sqrt(0.90)
display(R_D)
```

The life for both bearings is:

```python
L = 40e3*520*60
display(L)
```

That's a life of 1248 million revolutions. This life is supposed to be common to both the bearings. 

In order to select suitable bearings, we have to first determine the $C_{10}$ values. We can do that using the load-life-reliability relationship:

\begin{gather*}
    C_{10} = F_D \left[ \frac{x_D}{x_0 + (\theta - x_0)[\ln(1/R_D) ]^{1/b}}  \right]^{1/a}
\end{gather*}

Alternatively, we can use the reliability adjustment factor to obtain the reliability adjusted rating life:
\begin{gather*}
L_R = K_R L_{10}
\end{gather*}
followed by the use of the relation:
$C_{10} L_R^{1/a} = F L^{1/a}$, where $F$ and $L$ are the desired load and life for that bearing. 

If we look at the slides, $K_R = 0.64$ for $R_D = 0.95$ (given by both Timken and SKF). 

First, consider the angular contact ball bearing (bearing A):

```python
K_R = 0.64
L_10 = 1e6

L_R = K_R*L_10

af = 1.4
F_A = af*725*4.448  #1 lbf = 4.448 N
L_A = L  #calculated earlier

a = 3 # it is a ball bearing

C10_A = F_A*(L_A/L_R)**(1/a)
display(C10_A)
```

If we look at Table 11-2 in Shigley, we will find that the 02-series angular contact ball bearing that has a $C_{10}$ rating larger than the above value is one with a bore diameter $d = 65$ mm. 



Let us next consider the cylindrical roller bearing (bearing B):

```python
K_R = 0.64
L_10 = 1e6

L_R = K_R*L_10

af = 1.4
F_B = af*2235*4.448  #1 lbf = 4.448 N
L_B = L  #calculated earlier

a = 10/3 # it is a roller bearing

C10_B = F_B*(L/L_R)**(1/a)
display(C10_B)
```

If we look at Table 11-3 in Shigley, we will find the 03-series cylindrical roller bearing that has a $C_{10}$ rating larger than the above value is one with a bore diameter of $d = 65$ mm. 


So, as per the calculations:

The bore diameter for the angular contact ball bearing (A) should be 55 mm while that for the cylindrical roller (B) bearing should be 65 mm also. 

Let's check the actual reliabilities for the two selected bearings. 

We first determine the life ($L_R$) corresponding to a load equal to the selected $C_{10}$ value but on the reliability line corresponding to the design load and life. The ratio of this $L_R$ value to $L_{10}$ will give us the $K_R$ value from which we can estimate the reliability value. 

For bearing A:

```python
a = 3
C10_A_selected = 63.7e3 #selected C10 value
LR_A = (F_A/C10_A_selected)**a*L_A

KR_A_new = LR_A/L_10
display(KR_A_new)
```

Next, for bearing B:

```python
a = 10/3
C10_B_selected = 138e3 # selected C10 value
LR_B = (F_B/C10_B_selected)**a*L_B

KR_B_new = LR_B/L_10
display(KR_B_new)
```

Thus, for the selected bearing A, $K_R = 0.44$ while for the selected bearing B, $K_R = 0.595$. 

This $K_R$ is designated as $a_1$ by both Timken and SKF. 

From the tabulated values of $a_1 = K_R$ below, we note that the selected bearing A has a reliability greater than 97 percent while the selected bearing B has a reliability greater than 95 percent. 

Therefore, it is assured that the overall reliability will exceed 90 percent as required by the question.
