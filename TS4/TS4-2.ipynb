In the descriptions below, the bearing bush will be referred to as the "hole" while the crank pin will be referred to as the "shaft". 

```python
D = 20
d = D

FD = 0 # H means fundamental deviation of hole is 0
fd = -0.040 # e means fundamental deviation of hole is -0.040 mm (from Table 3.3a of Bhandari's book)

i = (0.45*D**(1/3) + 0.001*D)*1e-3  # same value if D is changed to d

IT6 = 10*i 
IT7 = 16*i

display(IT6, IT7)
```

Since the fundamental deviation of the hole is given by H, note that this fundamental deviation is 0 and it is equal to the lower deviation of the hole. The upper deviation is above this (corresponding to a larger hole). 

```python
Dmin = D + FD # Note that FD is zero; so Dmin = D
Dmax = Dmin + IT6
display(Dmin, Dmax)
```

Since the fundamental deviation of the shaft is given by e, note that it is the upper deviation that is equal to the fundamental deviation.The lower deviation is below this (corresponding to a smaller shaft). 

```python
dmax = d + fd  #Note that fd < 0; so dmax < d 
dmin = dmax - IT7  
display(dmax, dmin)
```

Maximum clearance corresponds to minimum shaft diameter AND maximum hole diameter. 

Minimum clearance corresponds to maximum shaft diameter AND minimum hole diameter. 

```python
max_clear = Dmax - dmin
min_clear = Dmin - dmax

display(max_clear, min_clear)
```
