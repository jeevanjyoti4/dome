In the descriptions below, the pump housing will be referred to as the "hole" while the valve seat will be referred to as the "shaft". 

```python
D = 20
d = D

FD = 0 # H means fundamental deviation of hole is 0
fd = 0.035 # s means fundamental deviation of hole is -0.040 mm (from Table 3.3a of Bhandari's book)

i = (0.45*D**(1/3) + 0.001*D)*1e-3  # same value if D is changed to d

IT7 = 16*i 
IT6 = 10*i

display(IT7, IT6)
```

Since the fundamental deviation of the hole is given by H, note that this fundamental deviation is 0 and it is equal to the lower deviation of the hole. The upper deviation is above this (corresponding to a larger hole). 

```python
Dmin = D + FD # Note that FD is zero; so Dmin = D
Dmax = Dmin + IT7
display(Dmin, Dmax)
```

Since the fundamental deviation of the shaft is given by s, note that it is the lower deviation that is equal to the fundamental deviation.The upper deviation is above this (corresponding to a larger shaft). 

```python
dmin = d + fd  #Note that fd > 0; so dmin > d 
dmax = dmin + IT6  
display(dmax, dmin)
```

Maximum interference corresponds to maximum shaft diameter AND minimum hole diameter. 

Minimum interference corresponds to minimum shaft diameter AND maximum hole diameter. 

```python
max_intf = dmax - Dmin
min_intf = dmin - Dmax

display(max_intf, min_intf)
```
