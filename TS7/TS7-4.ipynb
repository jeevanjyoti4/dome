```python
from math import pi, cos, sin
```

```python
phi = 20*pi/180
Z1 = 21
Z2 = 28
P = 3
```

```python
add = 1/P #addendum
ded = 1.25/P #dedendum
clearance = ded - add
p = pi/P #circular pitch
tooth_thickness = p/2

d1 = Z1/P # pitch circle diameter of pinion
d2 = Z2/P # pitch circle diameter of gear

d1b = d1*cos(phi) # base circle diameter of pinion
d2b = d2*cos(phi) # base circle diameter of gear

display(add, ded, clearance, p, tooth_thickness, d1b, d2b)
```

```python
from math import asin
```

```python
r1 = d1/2
r2 = d2/2
display(r1, r2)
```

```python
pb = p*cos(phi)  #base pitch
display(pb)
```

```python
from math import sqrt
```

```python
aP = sqrt( (r2+add)**2 - (r2*cos(phi))**2 ) - r2*sin(phi)
bP = sqrt( (r1+add)**2 - (r1*cos(phi))**2 ) - r1*sin(phi)
Lab = aP+bP
display(Lab)
```

```python
contact_ratio = Lab/pb
display(contact_ratio)
```
