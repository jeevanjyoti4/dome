```python
from math import pi
```

```python
m = 1e-3
Z = 16
H = 150 
n = 400

sigma = 150e6

d = m*Z

V = pi*d*(n/60)

Wt = H/V

Kv = (6.1+V)/6.1

Y = 0.296 # for 16 teeth on a 20 degree pressure angle system

F = Kv*Wt/(sigma*m*Y)
display(F)
```

The calculated answer is 10.6 mm. So the face width that has to be selected must be at least 11 mm.
