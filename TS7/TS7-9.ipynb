```python
from math import pi
```

```python
Z = 18
n = 900
H = 1.5e3

Y = 0.309 # for 18 teeth on a 20 degree pressure angle system
```

Since both the module and the face width are not given, we have to do trial and error. The target is to have a face width that is 3-5 times the circular pitch (for spur gears). 

We start with m = 1 mm, then try m = 10 mm, then m = 5 mm, and keep "zoning in", until we reach the following with m = 2 mm. 

```python
m = 2e-3

p = pi*m

d = m*Z

V = pi*d*(n/60)

Kv = (6.1 + V)/6.1 # cut teeth

Wt = H/V

sigma = 75e6

F = Kv*Wt/(sigma*m*Y)

display(F)

display(F/p) #in MPa
```

The desired face width can be 25 mm (just greater than the computed value of 24.38 mm).
