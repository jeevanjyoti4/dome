```python
phi_n = 20*pi/180
psi = 30*pi/180
Z = 18
n = 1200
mn = 3
```

```python
mt = mn/cos(psi) 

d = mt*Z

display(mt, d)
```

```python
V = pi*(d/1000)*n/60
display(V)
```

```python
Wt = 1000/V
display(Wt)
```

```python
Wr = Wt*tan(phi_n)/cos(psi)
Wa = Wt*tan(psi)
display(Wr, Wa)
```
