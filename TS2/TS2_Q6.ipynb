```python
from math import pi, sqrt
import sympy as sym
```

```python
t = 1.25e-3
d = 200e-3
r = d/2

Sy = 390e6  #From Table A-20 (yield strength)
Su = 470e6  #From Table A-20 (tensile strength)
```

```python
p = sym.symbols('p',positive=True)
```

```python
sigma_1 = p*r/(2*t)
sigma_2 = p*r/(2*t)
sigma_3 = -p 

sigma_dash = sym.sqrt(((sigma_1 - sigma_2)**2 + (sigma_2 - sigma_3)**2 + (sigma_3 - sigma_1)**2)/2 )

display(sigma_dash)
```

```python
pY_soln = sym.solve(sym.Eq(sigma_dash,Sy),p)[0]
display(pY_soln/1e6)
```

```python
pB_soln = sym.solve(sym.Eq(sigma_dash,Su),p)[0]
display(pB_soln/1e6)
```
