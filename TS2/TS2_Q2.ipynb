```python
Syt = 420
Syc = 525
```

```python
from math import sqrt
```

## Part (a)

There are no shear stresses. So the given normal stresses are themselves the principal stresses. 

$\sigma_A = 175$ and $\sigma_B = 105$

Here, $\sigma_A > \sigma_B > 0$. 

So, the failure criterion to be used is:
$\displaystyle \frac{\sigma_A}{S_{yt}} - \frac{0}{S_{yc}} = \frac{1}{n}$

```python
sigma_A = 175
n = Syt/sigma_A
display(n)
```

## Part (b)

Again, there are no shear stresses. So the given normal stresses are themselves the principal stresses. 

$\sigma_A = 105$ and $\sigma_B = -105$. 

Here, $\sigma_A > 0 > \sigma_B$. 

So, the failure criterion to be used is: $\displaystyle \frac{\sigma_A}{S_{yt}} - \frac{\sigma_B}{S_{yc}} = \frac{1}{n}$

```python
sigma_A = 105
sigma_B = -105

n = (sigma_A/Syt - sigma_B/Syc)**(-1)
display(n)
```

## Part (c)

This time there is shear stress. So, first the principal stresses have to be found. 

$\displaystyle \sigma_{A,B} = \frac{\sigma_x + \sigma_y}{2} \pm \sqrt{\left( \frac{\sigma_x - \sigma_y}{2} \right)^2 + \tau_{xy}^2}$

```python
sigma_x = 140
sigma_y = 0
tau_xy = -70

sigma_A = (sigma_x + sigma_y)/2 + sqrt( ((sigma_x - sigma_y)/2)**2 + tau_xy**2 )
sigma_B = (sigma_x + sigma_y)/2 - sqrt( ((sigma_x - sigma_y)/2)**2 + tau_xy**2 )
display(sigma_A, sigma_B)
```

Here, $\sigma_A > 0 > \sigma_B$

So, the failure criterion is:
$\displaystyle \frac{\sigma_A}{S_{yt}} - \frac{\sigma_B}{S_{yc}} = \frac{1}{n}$

```python
n = (sigma_A/Syt - sigma_B/Syc)**(-1)
display(n)
```
