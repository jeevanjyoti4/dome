```python
from math import pi, sqrt
```

```python
d = 15e-3
r = d/2
L = 100e-3

T = 25
F = 0.55e3
P = 4e3

Q = (4*r)/(3*pi)*(pi*r**2/2)
I = pi*r**4/4
J = pi*r**4/2
t = 2*r

Sy = 280e6  #from Table A-20 (Shighley 10th ed)
```

## For location A:

```python
tauA_xz =  T*r/J - F*Q/(I*t)
display(tauA_xz/1e6)

sigmaA_x = P/(pi*r**2)
display(sigmaA_x/1e6)
```

```python
sigmaA_dash = sqrt(sigmaA_x**2 + 3*tauA_xz**2)
nA = Sy/sigmaA_dash
display(nA)
```

## For location B:

```python
tauB_xy = T*r/J

sigmaB_x = (F*L)*r/I + P/(pi*r**2)

sigmaB_dash = sqrt(sigmaB_x**2 + 3*tauB_xy**2)

nB = Sy/sigmaB_dash
display(nB)
```
