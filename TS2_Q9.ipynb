```python
from math import sqrt
```

```python
I_di = 20e-3
I_do = 40e-3

O_di = 39.98e-3
O_do = 65e-3

I_ri = I_di/2
I_ro = I_do/2

O_ri = O_di/2
O_ro = O_do/2

delta = I_ro - O_ri

ri = I_ri
R = I_ro
ro = O_ro 

E = 207e9  # Shigley 10th ed, Table A-5
```

```python
p = E*delta/(2*R**3)*(ro**2 - R**2)*(R**2 - ri**2)/(ro**2 - ri**2)  # Shigley 10th ed, Eq 3-57
display(p/1e6)
```

## For inner tube:

```python
sigma_t = -p*(R**2 + I_ri**2)/(R**2 - I_ri**2)
sigma_r = -p

sigma_dash_I = sqrt(sigma_t**2 - sigma_t*sigma_r + sigma_r**2)
display(sigma_dash_I/1e6)
```

## For outer tube:

```python
sigma_t = p*(O_ro**2 + R**2)/(O_ro**2 - R**2)
sigma_r = -p

sigma_dash_O = sqrt(sigma_t**2 - sigma_t*sigma_r + sigma_r**2)
display(sigma_dash_O/1e6)
```
