```python
from math import sqrt
```

```python
do = 100e-3
t = 1.55e-3
di = do - 2*t

Sy = 345e6 #Table A-24 in Shigley 10th ed 

p = 3.5e6  #Equal to the pressure release valve value

sigma_1 = p*(di+t)/(2*t)  # Shigley 10th ed, Eq. 3-53 (formula for max. tangential stress)

sigma_2 = p*di/(4*t)   # Shigley 10th ed, Eq. 3-54

sigma_3  = -p

sigma_dash = sqrt( ( (sigma_1 - sigma_2)**2 + (sigma_2 - sigma_3)**2 + (sigma_3 - sigma_1)**2 )/2 )

n = Sy/sigma_dash
display(n)
```
