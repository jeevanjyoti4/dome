```python
l = 28e-3
dmax = 56e-3
bmin = 56.05e-3
W = 2.4e3
N = 900/60  #in rev/s
```

We will first calculate the Sommerfeld number: $S = \left( \dfrac{r}{c} \right)^2 \dfrac{\mu N}{P}$

```python
cmin = (bmin - dmax)/2
d = 56e-3
r = d/2

rc_ratio = r/cmin

P = W/(2*r*l)
```

To determine the dynamic viscosity $\mu$ of the SAE 40 lubricant at the given operating temperature of 65 degrees celsius, we use Fig. 12-3 in Shigley, 11th ed. 

From the figure, we read the value as $\mu = 30$ mPa$\cdot$s. 



Then, the Sommerfeld number is:

```python
mu = 30e-3
S = (rc_ratio)**2*(mu*N/P)
display(S)
```

The $l/d$ value is:

```python
l/d
```

Then, from Fig. 12-15, we have for the minimum film thickness:



The $h_0/c$ value corresponding to $S = 0.369$ and $l/d = 1/2$ is 0.44. 

Therefore, the $h_0$ value is:

```python
h0 = 0.44*cmin
display(h0)
```

So, $h_0 = 0.011$ mm. 

The power loss will be given by: $H = T(2\pi N)$. 

We note that $T = fWr$

Alternatively,


For the value of $f$ we use Fig. 12-17. From the figure, we read the value of $\left(\dfrac{r}{c}\right)f$ corresponding to $S= 0.369$ and $l/d = 1/2$ as 8. 



Then the value of $f$ is:

```python
f = 9/(rc_ratio)
display(f)
```

```python
T1 = f*W*r

T2 = 4*3.14**2*r**3*l*mu*N/cmin

display(T1, T2)
```

```python
T = T1
H = T*(2*3.14*N)
display(H)
```

We read $\dfrac{Q}{rcNl} = 4.7$ corresponding to $S=0.369$ and $l/d = 1/2$. 



Therefore, the value of $Q$ is:

```python
Q = 4.7*r*cmin*N*l
display(Q)
```

Therefore, the value of $Q_s$ is:

```python
Qs = 0.72*Q
display(Qs)
```
