```python
dmax = 25
bmin = 25.03

cmin = (bmin - dmax)/2 #minimum clearance
display(cmin)
```

```python
r = 25/2
rc_ratio = r/cmin  #radial clearance ratio
display(rc_ratio)
```

```python
N = 1100/60
display(N)
```

```python
ld_ratio = 1/2
d = 25
l = ld_ratio*d
W = 1.2e3
P = W/(l*d)
display(P) #in MPa
```

Next, we determine the Sommerfeld number

```python
mu = 55e-3
S = (rc_ratio)**2*(mu*N/(P*1e6))   #P is multiplied with 1e6 as it's in MPa
display(S)
```

To determine the minimum film thickness ($h_0$), we use Fig. 12-15 in Shigley (11th edition). The $h_0/c$ ratio comes out as 0.3. 



```python
h0 = 0.3*cmin
display(h0)
```

For power loss, we will use the relation: $H_{\rm loss} = T \omega$ = $T (2 \pi N)$. 

For $T$, we will use the relation: $T = f W r$. 

And, that means we need to determine $f$, the coefficient of friction. 

We use Fig. 12-17 from Shigley 11th edition:




We note that the ratio value of the ratio $fr/c$ corresponding to $S = 0.182$ and $l/d=1/2$ is 5.4.

```python
f = 5.4/(rc_ratio)
display(f)
```

```python
T = f*W*r/1e3  #dividing by 1e3 because r is in mm
H_loss = T*2*3.14*N
display(H_loss)
```

Finally, for the side flow we will use Figs 12-18 and 12-19 from Shigley 11th edition. 

From Fig. 12-18 below we note that the value of the ratio $Q/(rcNl) = 5.1$ corresponding to $S = 0.182$ and $l/d = 1/2$.  



Therefore, the value of $Q$ is:

```python
Q = 5.1*r*cmin*N*l
display(Q)
```

From Fig. 12-19, we note that $Q_s/Q = 0.81$ corresponding to $S = 0.182$ and $l/d = 1/2$. 



Therefore, the value of the side flow $Q_s$ is:

```python
Qs = Q*0.81
display(Qs) 
```
