```python
import sympy as sym
```

```python
di = 42e-3
do = 50e-3
L = 150e-3

ri = di/2
ro = do/2

n = 2

Sy = 490 # Shigley 10th ed, Table A-20
```

```python
p = sym.symbols('p', positive=True)
```

```python
sigma_t = p*(ro**2 + ri**2)/(ro**2 - ri**2)
sigma_l = p*ri**2/(ro**2 - ri**2)
sigma_r = -p
display(sigma_t, sigma_l, sigma_r)
```

## For Part (a): Maximum shear stress theory:

```python
tau_max = (sigma_t - sigma_r)/2
display(tau_max)
MSS_eqn = sym.Eq(tau_max, Sy/(2*n))

pMSS_soln = sym.solve(MSS_eqn, p)[0]
display(pMSS_soln)
```

## For Part (b): Distortion energy theory

```python
sigma_dash = sym.sqrt( ( (sigma_t - sigma_l)**2 + (sigma_l - sigma_r)**2 + (sigma_r - sigma_t)**2 )/2 )

DE_eqn = sym.Eq(sigma_dash, Sy/n)

pDE_soln = sym.solve(DE_eqn, p)[0]
display(pDE_soln)
```
