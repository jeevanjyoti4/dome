```python
from math import sqrt
```

## Part (a) 
Since there are no shear stresses, $\sigma_A = \sigma_x$ and $\sigma_B = \sigma_y$. But, $0$ is also a principal stress. Here $\sigma_A > \sigma_B > 0$. 

So,
According to MSS: $\sigma_A - 0 = \dfrac{S_y}{n}$. 

According to DE: $\sigma^\prime = (\sigma_A^2 - \sigma_A \sigma_B + \sigma_B^2)^{1/2} = \dfrac{S_y}{n}$


```python
sigma_A = 100
sigma_B = 100

n_MSS = Sy/(sigma_A - 0)


sigma_dash = sqrt(sigma_A**2 - sigma_A*sigma_B + sigma_B**2)

n_DE = Sy/sigma_dash

display(n_MSS, n_DE)
```

## Part (c)

This time there is shear stress. So, first the principal stresses have to be found. 

$\displaystyle \sigma_{A,B} = \frac{\sigma_x + \sigma_y}{2} \pm \sqrt{\left( \frac{\sigma_x - \sigma_y}{2} \right)^2 + \tau_{xy}^2}$

```python
sigma_x = 100
sigma_y = 0
tau_xy = -75

sigma_A = (sigma_x + sigma_y)/2 + sqrt( ((sigma_x - sigma_y)/2)**2 + tau_xy**2 )
sigma_B = (sigma_x + sigma_y)/2 - sqrt( ((sigma_x - sigma_y)/2)**2 + tau_xy**2 )
display(sigma_A, sigma_B)
```

Here, $\sigma_A > 0 > \sigma_B$

So,
According to MSS: $\sigma_A - \sigma_B = \dfrac{S_y}{n}$. 

According to DE: $\sigma^\prime = (\sigma_A^2 - \sigma_A \sigma_B + \sigma_B^2)^{1/2} = \dfrac{S_y}{n}$

```python
n_MSS = Sy/(sigma_A - sigma_B)

sigma_dash = sqrt(sigma_A**2 - sigma_A*sigma_B + sigma_B**2)

n_DE = Sy/sigma_dash

display(n_MSS, n_DE)
```
